### Data Wrangling

```python
# Check for any missing values
df_price.isnull().sum()
```

```python
# It's hard to see where any of the null values are
# Create a new subset with only null values

df_na = df_price[df_price['Metro'].isnull() == True]
df_na
```

```python
# Since there are plenty of rows to work with, omitting these null values will make analysis less complicated

df_price = df_price[df_price['Metro'].isnull() == False]
df_price
```

```python
# A quick glance at excel shows there are multiple properties that are null between the beginning of the data
# (11/1/2010) and up to 3/1/2012
# I chose to omit just these specific dates

df_price = df_price.drop(['November 2010','December 2010','January 2011','February 2011','March 2011','April 2011',
                          'May 2011','June 2011','July 2011','August 2011','September 2011','October 2011','November 2011',
                          'December 2011','January 2012','February 2012','March 2012'
              ], axis=1)

df_price.iloc[:,:10]
```

```python
# Considering the scale of the dataset, it will be hard to create visualiztions that are processor-friendly
# For the purposes of this project, I will only be using the top 500 cities and do analysis on those cities

df_top_500_cities = df_price.head(500)
df_top_500_cities
```

```python
# Append all month/year into a row corresponding to each city

df_transposed = pd.melt(df_top_500_cities, id_vars=['City Code','City','Metro','County','State','Population Rank'],
                       var_name = 'Month/Year', value_name='Rent Price')
```
